# Seção 3: Treinando o Modelo house_prices_model com Amazon SageMaker

Nesta seção, vamos focar no treinamento do modelo house_prices_model, um modelo de regressão linear para previsão de preços de casas. Vamos utilizar o Amazon SageMaker para treinar, avaliar e implantar o modelo.

## 3.1 Preparação dos Dados

Certifique-se de que você já carregou o conjunto de dados house_prices.csv para o seu bucket S3, conforme explicado na Seção 2.

## 3.2 Acessando o Jupyter Notebook no SageMaker

1. Acesse o Console da AWS e vá para o serviço Amazon SageMaker.

2. Clique em “Notebook instances”.

3. Localize sua instância e clique em “Open Jupyter”.

## 3.3 Criando o Script de Treinamento no SageMaker

Passo 1: Criar um Novo Notebook Jupyter

1. No Jupyter Notebook, clique em “New” e escolha um kernel Python.

2. Dê um nome ao seu notebook, como HousePricesModelTraining.

Passo 2: Importar Bibliotecas e Carregar Dados

- Importe as bibliotecas necessárias e carregue o conjunto de dados do S3.

``` python
import pandas as pd
import joblib
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import boto3
import sagemaker
from sagemaker import get_execution_role

# Configurar a sessão do SageMaker e o papel
sagemaker_session = sagemaker.Session()
role = get_execution_role()

# Baixar o conjunto de dados do S3
bucket_name = 'seu-bucket'
data_key = 'house_prices.csv'
data_location = 's3://{}/{}'.format(bucket_name, data_key)

data = pd.read_csv(data_location)
````

Substitua 'seu-bucket' e 'house_prices.csv' com o nome do seu bucket S3 e o caminho do seu arquivo de dados, respectivamente.

Passo 3: Preparar os Dados

``` python
# Separar as variáveis independes e dependentes
X = data.drop('Price', axis=1)
Y = data['Price']

# Dividir os dados em conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

Passo 4: Treinar o Modelo

``` python
# Criar e treinar o modelo
model = LinearRegression()
model.fit(X_train, y_train)

# Avaliar o modelo
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
```

Passo 5: Salvar o Modelo Treinado e fazer upload no S3
``` python
# Salvar o modelo localmente e fazer upload no S3
import boto3

joblib.dump(model, 'model.pkl')

s3 = boto3.client('s3')
bucket_name = 'your-s3-bucket-name'
object_name = 'path/to/your/model.pkl'
s3.upload_file('model.pkl', bucket_name, object_name)
```

O SageMaker automaticamente carregará os artefatos salvos neste diretório para o S3 após a conclusão do job de treinamento.


Com estas etapas, você terá treinado e avaliado o modelo house_prices_model usando os recursos do Amazon SageMaker no Free Tier. Na próxima seção, aprenderemos como implantar este modelo e realizar previsões.


# Seção 4: Registrando um Contêiner Docker no Amazon ECR

## 4.1 Introdução

Nesta seção, você aprenderá a transferir uma imagem Docker do Docker Hub para o Amazon Elastic Container Registry (ECR). Isso permitirá que você use esta imagem com os serviços da AWS, incluindo o SageMaker.

## 4.2 Pré-requisitos

1. Uma conta AWS

2. AWS CLI configurado em sua máquina local

3. Docker instalado em sua máquina local

## 4.3. Instalando o AWS Command Line Interface (CLI)

Antes de começarmos a trabalhar com o Amazon Elastic Container Registry (ECR) e o Docker, precisamos garantir que o AWS Command Line Interface (CLI) esteja instalado em nosso sistema. O AWS CLI é uma ferramenta de linha de comando que permite interagir com os serviços da AWS diretamente do terminal. Aqui está como você pode instalá-lo:

### Windows

1. Baixe o instalador: Vá até a [página de download do AWS CLI](https://aws.amazon.com/cli/) e baixe o instalador apropriado para a sua versão do Windows (32 bits ou 64 bits).

2. Execute o instalador: Uma vez baixado, execute o instalador e siga as instruções na tela.

3. Verifique a instalação: Abra um terminal e digite aws --version para verificar se o AWS CLI foi instalado corretamente. Você deverá ver a versão instalada como resposta.

### macOS

1. Baixe o pacote de instalação: Vá até a página de download do AWS CLI e baixe o pacote de instalação.

2. Instale usando o pacote: Abra o pacote baixado e siga as instruções na tela.

3. Verifique a instalação: Abra um terminal e digite aws --version para verificar se o AWS CLI foi instalado corretamente. Você deverá ver a versão instalada como resposta.

### Linux

1. Baixe o arquivo de instalação: Abra um terminal e digite:

``` sh
curl "https://d1vvhvl2y92vvt.cloudfront.net/awscli-exe-linux-x86_64.zip" -o "awscliv2.zip"
```

2. Descompacte o arquivo:

``` sh
unzip awscliv2.zip
```

3. Execute o script de instalação:

``` sh
sudo ./aws/install
```

Verifique a instalação: Digite aws --version para verificar se o AWS CLI foi instalado corretamente. Você deverá ver a versão instalada como resposta.

Depois de instalar o AWS CLI, você também precisará configurá-lo com suas credenciais da AWS. Você pode fazer isso executando aws configure e fornecendo seu ID de chave de acesso, chave de acesso secreta, região padrão e formato de saída desejado.

Agora que o AWS CLI está instalado e configurado, você pode seguir os passos descritos na Seção 4 para registrar um contêiner Docker no Amazon ECR.

## 4.4 Passos

### Passo 1: Criar um Repositório ECR

1. Vá até o console AWS e selecione o serviço Elastic Container Registry.

2. Clique em “Create repository”.

3. Dê um nome ao seu repositório e clique em “Create repository”.

4. Anote o URI do repositório, você precisará dele mais tarde.


### Passo 2: Autenticar o Docker ao seu Registro ECR

1. Execute o seguinte comando no seu terminal para autenticar o Docker ao seu registro ECR.

```sh
aws ecr get-login-password --region sua-região | docker login --username AWS --password-stdin seu-id-de-conta.dkr.ecr.sua-região.amazonaws.com
```

Substitua sua-região, seu-id-de-conta com a sua região AWS e o ID da sua conta AWS, respectivamente.

### Passo 3: Puxar a Imagem Docker do Docker Hub

``` sh
docker pull nome-do-usuário/nome-da-imagem:tag
```

Substitua nome-do-usuário, nome-da-imagem, e tag com os detalhes apropriados da imagem que você deseja puxar do Docker Hub.

### Passo 4: Marcar a Imagem Docker

``` sh
docker tag nome-do-usuário/nome-da-imagem:tag seu-id-de-conta.dkr.ecr.sua-região.amazonaws.com/nome-do-repositório:tag
```

### Passo 5: Enviar a Imagem para o ECR

``` sh
docker push seu-id-de-conta.dkr.ecr.sua-região.amazonaws.com/nome-do-repositório:tag
```

4.4 Conclusão
Agora você tem sua imagem Docker hospedada no Amazon ECR e está pronta para ser usada com os serviços AWS, incluindo o Amazon SageMaker. No próximo passo, você aprenderá como utilizar essa imagem para criar e implantar modelos de machine learning.

Lembre-se de substituir todos os placeholders (como nome-do-usuário, nome-da-imagem, tag, sua-região, seu-id-de-conta, e nome-do-repositório) com as informações reais do seu ambiente.

# Seção 5: Conclusão e Próximos Passos

Parabéns por concluir esta aula importante na sua jornada de aprendizado em machine learning com a AWS! Durante esta aula, você teve a oportunidade de explorar e utilizar recursos poderosos da AWS e do Amazon SageMaker para criar e gerenciar modelos de machine learning na nuvem.

## Principais Pontos:

1- Introdução à AWS e ao SageMaker: Você começou a aula com uma introdução detalhada sobre a AWS, aprendendo como criar uma conta no Free Tier e como navegar no console da AWS. Também explorou o Amazon SageMaker, uma ferramenta incrivelmente poderosa para criar, treinar e implantar modelos de machine learning.

2. Treinamento de Modelos na Nuvem: Utilizando o SageMaker, você treinou o modelo house_prices_model diretamente na nuvem, aproveitando os recursos escaláveis e flexíveis que a AWS oferece.

3. Registro de Containers no ECR: Aprendeu o processo de registrar um container Docker no Amazon Elastic Container Registry (ECR), preparando o terreno para implantações futuras.

4. AWS CLI e Gerenciamento de Permissões: Adquiriu prática com a AWS CLI e enfrentou desafios reais de gerenciamento de permissões, reforçando conceitos essenciais de segurança.

## Na Próxima Aula:

Com todas essas habilidades adquiridas, você está bem preparado para a próxima aula, onde:

1. Criará Funções Lambda: Para interagir de forma programática com seu modelo de machine learning.

2. Configurará o API Gateway: Para disponibilizar seu modelo como um serviço web acessível.

Integrará Lambda, API Gateway e ECR: Para criar uma solução completa de machine learning como serviço, permitindo que usuários façam previsões através de uma API RESTful.

Esta próxima etapa é crucial para disponibilizar seu modelo para o mundo, transformando-o em uma ferramenta prática e acessível. Estamos ansiosos para guiá-lo através destes conceitos avançados e ver o incrível progresso que você fará!
